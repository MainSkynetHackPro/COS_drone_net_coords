```python
import json
from matplotlib import pyplot as plt
import copy
%matplotlib inline


DATASET_FILE = 'data_4/test.json'
DATASET_FILE_EQUALIZED = 'data_4/test_eq.json'
```

```python
with open(DATASET_FILE) as f:
    data = json.loads(f.read())
    
len(data)
```

```python
for i in data:
    i['copy'] = False
```

```python
data
```

```python
drones = list([copy.copy(i) for i in filter(lambda x: x['label'] == 'drone', data)])
bg = list(filter(lambda x:x['label'] == 'bg', data))

drones_count = len(drones)
bg_count = len(bg)
```

```python
p = plt.bar(['drones', 'bg'], [drones_count, bg_count])
```

```python
for i in drones:
    i['copy'] = True
```

```python
diff = bg_count - drones_count
data_new = data
for i in range(0, diff, drones_count):
    data_new += drones

drones = list(filter(lambda x: x['label'] == 'drone', data_new))
bg = list(filter(lambda x:x['label'] == 'bg', data_new))

drones_count = len(drones)
bg_count = len(bg)

p = plt.bar(['drones', 'bg'], [drones_count, bg_count])
```

```python
with open(DATASET_FILE_EQUALIZED, 'w') as f:
    f.write(json.dumps(data_new))
```
